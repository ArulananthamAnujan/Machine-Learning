### **The Implementation of the Alex-Net Archietecture**:Alexnet.png

We can address this in the simple terms that 
  
> For the Alexnet Diagram

    1.Input Image ----> 227x227x3
    2.Convolutional Layer  ---->96 11x11 
    3.Maxpooling Layer ---->96 3x3
    4.Convolutional Layer  ---->256 3x3
    5.Maxpooling Layer ---->256 3x3
    5.Convolutional Layer  ---->384 3x3
    6.Convolutional Layer  ---->384 3x3
    7.Convolutional Layer  ---->256 3x3
    6.Maxpooling Layer ---->256 3x3
    7.Dense Layer Network with First Layer with 4096 Neurons
    7.Dense Layer Network with Second Layer with 4096 Neurons
    7.Dense Layer Network with Third Layer with 1000 Neurons ----->Classification Output



Alexnet-Dimension.png

```python
#Importing the Necessary Files
import keras
import numpy as np
from keras.models import Sequential
from keras.layers import Dense,Activation,Dropout,Flatten,Conv2D,MaxPooling2D,BatchNormalization

#the size of the image
image_size=(227,277,3)

np.random.seed(1000)# the instance of the empty model

#Start a Model to train process
model=Sequential()

#First Convolutional_Layer
model.add(Conv2D(filters=96,input_shape=image_size,kernel_size=11,strides=4,padding='valid'))
model.add(Activation('relu'))

#Followed by Max-Pooling Layer
model.add(MaxPooling2D(pool_size=3,strides=2,padding='valid'))

#Second Convolutional_Layer
model.add(Conv2D(filters=256,kernel_size=5,strides=1,padding='valid'))
model.add(Activation('relu'))

#Followed by Max-Pooling Layer
model.add(MaxPooling2D(pool_size=3,strides=2,padding='valid'))

#Third Convolutional_Layer
model.add(Conv2D(filters=384,kernel_size=3,strides=1,padding='valid'))
model.add(Activation('relu'))

#Fourth Convolutional_Layer
model.add(Conv2D(filters=384,kernel_size=3,strides=1,padding='valid'))
model.add(Activation('relu'))

#Fifth Convolutional_Layer
model.add(Conv2D(filters=256,kernel_size=3,strides=1,padding='valid'))
model.add(Activation('relu'))

#Followed by Pooling Layer
model.add(MaxPooling2D(pool_size=3,strides=2,padding='valid'))

#Flatten the layers 
model.add(Flatten())

#Implementation of the Dense Layer
model.add(Dense(4096,input_shape=(227*227*3,)))
model.add(Dropout(0.4))

#implement the Fully Connected Layer
model.add(Dense(4096))
model.add(Dropout(0.4))

#The Last Layer
model.add(Dense(1000))


model.summary()

model.compile(optimizer='adam',loss='catergorical_crossentropy',metrics=['accuracy'])
```
